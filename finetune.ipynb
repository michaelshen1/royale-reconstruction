```python
import ultralytics
from ultralytics import YOLO
import torch
import os
```

```python
# hyperparameters
EPOCHS = 100
IMG_SIZE = 896

os. environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
model = YOLO('yolo26n.pt')

model.to('cuda' if torch.cuda.is_available() else 'cpu')

# Feed in low num of epochs to maek sure it works
EPOCHS = 5

# Have different name = ___ every time running this, or else will override
results = model.train(data = './data.yaml', epochs = EPOCHS, imgsz = IMG_SIZE, batch = 8, project = "./models",
                      name = "clash_yolo_v1", rect = True, mosaic = 0.0, scale = 0.0, degrees = 0.0, perspective = 0.0,
                      amp = True, exist_ok = True, save_period = 5)
```
